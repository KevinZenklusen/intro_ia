```python
from gensim.models import Word2Vec
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Frases de ejemplo
sentences = [
    ["el", "perro", "corre", "rápido"],
    ["el", "gato", "salta", "alto"],
    ["el", "ratón", "come", "queso"],
    ["el", "cuchillo", "corta", "bien"],
    ["el", "tenedor", "pincha", "bien"],
    ["los", "gatos", "duermen", "mucho"],
    ["los", "perros", "ladran", "fuerte"]
]

# Entrenamiento del modelo Word2Vec (Bag of Words)
model_1 = Word2Vec(sentences, vector_size=10, window=2, min_count=1, sg=0)

# Palabras presentes en el modelo
words = list(model_1.wv.index_to_key)

# Vectorización de una frase
phrase = ["el", "gato", "salta"]
vectorized_phrase = model_1.wv[phrase]
print(f"Frase: {phrase}")
print(f"Vectorización de la frase: {vectorized_phrase}")

# Creamos una lista con las parabras de las frases (Sin repeticiones)
single_word_list = list(set([element for sublist in sentences for element in sublist]))

# Comparamos un vector consigo mismo
vector_1 = model_1.wv['gatos']
vector_2 = model_1.wv['gato']
similarity = cosine_similarity(np.array(vector_1).reshape(1, -1), np.array(vector_2).reshape(1, -1))
print(similarity[0][0])

for word in single_word_list:
    vector_2 = model_1.wv[word]
    # Calculando la distancia coseno entre los vectores
    similarity = cosine_similarity(np.array(vector_1).reshape(1, -1), np.array(vector_2).reshape(1, -1))
    print(f"Similaridad entre gatos y '{word}': {similarity[0][0]}")
```
